```python
tokenized_dataset = tokenizer(
    raw_datasets["train"]["sentence1"],
    raw_datasets["train"]["sentence2"],
    padding=True,
    truncation=True,
)
```

```python
tokenized_dataset.keys()
```

```python
count_none_zero_elements = {'x':[],'non_zero_count':[]}
for i in range(len(tokenized_dataset['input_ids'])):

  non_zero_count = 0
  for j in range(len(tokenized_dataset['input_ids'][i])):

    if tokenized_dataset['input_ids'][i][j] != 0:
      non_zero_count = non_zero_count + 1
    count_none_zero_elements['x'].append(i)
    count_none_zero_elements['non_zero_count'].append(non_zero_count)
```

```python
max_non_zero_count=max(count_none_zero_elements['non_zero_count'])
max_index = count_none_zero_elements['non_zero_count'].index(max_non_zero_count)
corresponding_x = count_none_zero_elements['x'][max_index]

# Print the results
print(f"The maximum non-zero count is: {max_non_zero_count}")
print(f"This occurs at x = {corresponding_x}")
tokenized_dataset['input_ids'][corresponding_x]
```

```python
tokenizer.convert_ids_to_tokens(tokenized_dataset['input_ids'][corresponding_x])
```

```python
def tokenize_function(example):
    return tokenizer(example["sentence1"], example["sentence2"], truncation=True)
```

```python
tokenized_datasets = raw_datasets.map(tokenize_function, batched=True)
tokenized_datasets
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
samples = tokenized_datasets["train"][:8]
samples = {k: v for k, v in samples.items() if k not in ["idx", "sentence1", "sentence2"]}
[len(x) for x in samples["input_ids"]]
```
